```python
def extract_single_drone(s):
    lines = s.split('\n')
    data = []
    for line in lines[2:]:
        curr = {'Gen':0, 'Counter':0, 'Points':0}
        line = line.split('!')[-1]
        params = line.split(',')
        for param in params:
            nums = param.split(': ')
            if len(nums) > 1:
                nums[0] = nums[0].lstrip()
                curr[nums[0]] = nums[1]
        data.append(curr)
    return data
```

```python
def extract_multi_drone(s):
    lines = s.split('\n')
    data = []
    for line in lines[2:-1]:
        line = line[10:]
        comma1 = line.index(',')
        gen = line[5:comma1]
        comma2 = line.index(',', comma1+1)
        count = line[comma1+11:comma2]
        points_idx_s = line.index('Points') + 8
        points_idx_e = line.index('Total') - 2
        points_list = line[points_idx_s:points_idx_e]
        if len(points_list) == 2:
            alive = 0
        elif len(points_list) == 3:
            alive = 1
        else:
            alive = points_list.count(',') + 1
        total = line[line.index('Total')+14:-4]
        data.append({'Gen': gen, 'Counter': count, 'Points': total, 'Num Alive': alive})
    return data
        
print(extract_multi_drone(ds4))
```

```python
def extract_info(string):
    lines = string.split('\n')
    data = {}
    for line in lines:
        params = line.split(',')
        # print(params)
        for param in params:
            nums = param.split(': ')
            if len(nums) > 1:
                if nums[0] not in data:
                    data[nums[0]] = []
                data[nums[0]].append(float(nums[1]))
    return data
```
